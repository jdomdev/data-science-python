5 - Escribir un programa para gestionar un listín telefónico con los nombres y los teléfonos de los clientes de una empresa. El programa incorporar funciones crear el fichero con el listín si no existe, para consultar el teléfono de un cliente, añadir el teléfono de un nuevo cliente y eliminar el teléfono de un cliente. El listín debe estar guardado en el fichero de texto listin.txt donde el nombre del cliente y su teléfono deben aparecer separados por comas y cada cliente en una línea distinta.

```python
def get_phone(file, client):
    '''
    Función que devuelve el teléfono de un cliente de un fichero dado.
    Parámetros:
        file: Es un fichero con los nombres y teléfonos de clientes.
        client: Es una cadena con el nombre del cliente.
    Devuelve:
        El teléfono del cliente guardado en el fichero o un mensaje de error si el teléfono o el fichero no existe.
    '''
     # Iniciamos bloque try-except-else
    try: 
        # Abrimos argumento fichero para leer información 
        f = open(file, 'r')
    # Capturamos excepción Fichero no encontrado 
    except FileNotFoundError:
        # Devolvemos cadena informando
        return('¡El fichero ' + file + ' no existe!\n')
    # Si no salta excepción
    else:
        # Leemos las líneas del fichero
        directory = f.readlines()
        # Cerramos fichero
        f.close()
        # Creamos un diccionario con las líneas del diccionario
        directory = dict([tuple(line.split(',')) for line in directory])
        # Si el argumento cliente se encuentra en el diccionario
        if client in directory:
            # Devolvemos el cliente
            return directory[client]
        # Si no se encuentra, devolvemos cadena informando
        else:
            return('¡El cliente ' + client + ' no existe!\n')


def add_phone(file, client, telf):
    '''
    Función que añade el teléfono de un cliente de un fichero dado.
    Parámetros:
        file: Es un fichero con los nombres y teléfonos de clientes.
        client: Es una cadena con el nombre del cliente.
        telf: Es una cadena con el teléfono del cliente.
    Devuelve:
        Un mesaje informando sobre si el teléfono se ha añadido o ha habido algún problema.
    '''
    # Iniciamos bloque try-except-else
    try: 
        # Abrimos argumento fichero para añadir información 
        f = open(file, 'a')
    # Capturamos excepción Fichero no encontrado    
    except FileNotFoundError:
        # Devolvemos cadena informando
        return('¡El fichero ' + file + ' no existe!\n')
    # Si no salta excepción
    else:
        # Escribimos en fichero cliente y teléfono
        f.write(client + ',' + telf + '\n')
        # Cerramos fichero
        f.close()
        # Devolvemos cadena informando
        return('El teléfono se ha añadido.\n')
    
# Definición de la función que elimina un número y recibe dos argumentos
def remove_phone(file, client):
    '''
    Función que elimina el teléfono de un cliente de un fichero dado.
    Parámetros:
        file: Es un fichero con los nombres y teléfonos de clientes.
        client: Es una cadena con el nombre del cliente.
    Devuelve:
        Un mesaje informando sobre si el teléfono se ha borrado o ha habido algún problema.
    '''
    # Iniciamos bloque try-except-else
    try: 
        # Abrimos el argumento fichero en modo lectura
        f = open(file, 'r')
    # Capturamos la excepción 'Error de fichero no encontrado'
    except FileNotFoundError:
        # Devolvemos una cadena informando del error
        return('¡El fichero ' + file + ' no existe!\n')
    # Si no salta excepción
    else:
        # Leemos las líneas del fichero
        directory = f.readlines()
        # Cerramos el fichero
        f.close()
        # Recorremos las lísta de líneas que convertimos, a su vez, en listas con
        # las cadenas que contienen divididas por comas, y  hacemos una tupla de esa lista
        # de los elementos de la línea, para acabar formando un diccionario con esas tuplas
        directory = dict([tuple(line.split(',')) for line in directory])
        # Si el argumento 'client' se encuentra en el diccionario recién creado
        if client in directory:
            # Eliminamos ese cliente del directorio
            del directory[client]
            # Abrimos el fichero en modo escritura
            f = open(file, 'w')
            # Bucle con dos variables que recorre el diccionario
            for name, telf in directory.items():
                # Escribimos en el fichero el nombre y el teléfono del diccionario
                f.write(name + ',' + telf)
            # Cerramos el fichero    
            f.close()
            # Devolvemos cadena informando del éxito de la operación de eliminación
            return ('¡El cliente se ha borrado!\n')
        # Si el cliente no se encuentra en el diccionario
        else:
            # Devolvemos cadena informando de que no existe el elemento buscado
            return('¡El cliente ' + client + ' no existe!\n')

# Definición de la función que crea el directorio y recibe un argumento
def create_directory(file):
    '''
    Función que crea un fichero vacío para guardar un listín telefónico.
    Parámetros:
        file: Es un fichero.
    Devuelve:
        Un mesaje informando sobre si el fichero se ha creado correctamente o no.
    '''
    # Importamos la biblioteca 'os'
    import os
    # Comprobamos que la ruta que se recibe contiene un archivo
    # Si se cumple la condición
    if os.path.isfile(file):
        # Se recoge de la introducción por teclado si se quiere sobrescribir el fichero
        answer = input('El fichero ' + file + ' ya existe. ¿Desea vaciarlo (S/N)? ')
        # Si la respuesta es negativa
        if answer == 'N': 
            # Se devuelve una cadena informando
            return 'No se ha creado el fichero porque ya existe.\n'
    # Si no se cumple la condición, se abre un fichero en modo escritura con el 
    # argumento introducido en la función
    f = open(file, 'w')
    # Cerramos el fichero
    f.close()
    # Devolvemos una cadena informando
    return 'Se ha creado el fichero.\n'
     
# Definición de la función que nuestra las operaciones disponibles
def menu():
    '''
    Función que presenta un menú con las operaciones disponibles sobre un listín telefónico y devuelve la opción seleccionada por el usuario.
    Devuelve:
        La opción seleccionada por el usuario.
    '''
    # Imprimimos las diferentes opciones
    print('Gesión del listín telefónico')
    print('============================')
    print('1 - Consultar un teléfono')
    print('2 - Añadir un teléfono')
    print('3 - Eliminar un teléfono')
    print('4 - Crear el listín')
    print('0 - Terminar')
    # Recogemos la opción elegida
    option = input('Introduzca el número de la opción deseada: ')
    # Devolvemos la opción elegida
    return option

# Primera función que se ejecuta
def directory():
    '''
    Función que lanza la aplicación para la gestión del listín telefónico.
    '''
    # Variable que recoge el path al fichero que recoge los datos
    file = '../data/listin.txt' 
    # Iniciamos bucle infinito
    while True:
        # Llamamos a función que imprime el menú de opciones
        # y recogemos la devolución
        option = menu()
        # Iniciamos condicional if-elif-else
        # Si la función menu() devuelve cadena '1'
        if option == '1':
            # Recogemos la cadena introducida por teclado(nombre)
            name = input('Introduce el nombre del cliente: ')
            # Imprimimos la llamada a función que busca contacto con argumentos 
            # fichero(cadena del path) y nombre(cadena)
            print(get_phone(file, name))
        # Si la función menu() devuelve cadena '2'
        elif option == '2':
            # Recogemos la cadena introducida por teclado(nombre)
            name = input('Introduce el nombre del cliente: ')
            # Recogemos la cadena introducida por teclado(teléfono)
            telf = input('Introduce el teléfono del cliente: ')
            # Imprimimos la llamada a función que añade un contacto con argumentos 
            # fichero(cadena del path), nombre(cadena) y teléfono(cadena)
            print(add_phone(file, name, telf))
        # Si la función menu() devuelve cadena '3'
        elif option == '3':
            # Recogemos la cadena introducida por teclado(nombre)
            name = input('Introduce el nombre del cliente: ')
            # Imprimimos la llamada a función que elimina un contacto con argumentos 
            # fichero(cadena del path) y nombre(cadena)
            print(remove_phone(file, name))
        # Si la función menu() devuelve cadena '4'
        elif option == '4':
            # Imprimimos la llamada a función que crea el listín con argumento 
            # fichero(cadena del path)
            print(create_directory(file))
        # Si la función menu() no devuelve ninguna de las anteriores
        else:
            # Rompemos el bucle y finalizamos la ejecución del progrmama
            break
    return
# Inicio del programa
directory()
```

```python
import os

def get_phone(file, client):
    try:
        with open(file, 'r') as f:
            directory = dict(line.strip().split(',') for line in f)
    except FileNotFoundError:
        return f'¡El fichero {file} no existe!\n'
    
    return directory.get(client, f'¡El cliente {client} no existe!\n')

def add_phone(file, client, telf):
    try:
        with open(file, 'a') as f:
            f.write(f'{client},{telf}\n')
        return 'El teléfono se ha añadido.\n'
    except FileNotFoundError:
        return f'¡El fichero {file} no existe!\n'

def remove_phone(file, client):
    try:
        with open(file, 'r') as f:
            directory = dict(line.strip().split(',') for line in f)
    except FileNotFoundError:
        return f'¡El fichero {file} no existe!\n'
    
    if client in directory:
        del directory[client]
        with open(file, 'w') as f:
            for name, telf in directory.items():
                f.write(f'{name},{telf}\n')
        return '¡El cliente se ha borrado!\n'
    return f'¡El cliente {client} no existe!\n'

def create_directory(file):
    if os.path.isfile(file):
        if input(f'El fichero {file} ya existe. ¿Desea vaciarlo (S/N)? ') == 'N':
            return 'No se ha creado el fichero porque ya existe.\n'
    with open(file, 'w'):
        pass
    return 'Se ha creado el fichero.\n'

def menu():
    print('Gesión del listín telefónico\n============================')
    print('1 - Consultar un teléfono\n2 - Añadir un teléfono\n3 - Eliminar un teléfono\n4 - Crear el listín\n0 - Terminar')
    return input('Introduzca el número de la opción deseada: ')

def directory():
    file = '../files/listin2.txt'
    while True:
        option = menu()
        if option == '1':
            print(get_phone(file, input('Introduce el nombre del cliente: ')))
        elif option == '2':
            print(add_phone(file, input('Introduce el nombre del cliente: '), input('Introduce el teléfono del cliente: ')))
        elif option == '3':
            print(remove_phone(file, input('Introduce el nombre del cliente: ')))
        elif option == '4':
            print(create_directory(file))
        else:
            break

directory()
```

6 - El fichero [cotizacion.csv](https://aprendeconalf.es/docencia/python/ejercicios/soluciones/ficheros/cotizacion.csv) contiene las cotizaciones de las empresas del IBEX35 con las siguientes columnas: Nombre (nombre de la empresa), Final (precio de la acción al cierre de bolsa), Máximo (precio máximo de la acción durante la jornada), Mínimo (precio mínimo de la acción durante la jornada), Volumen (Volumen al cierre de bolsa), Efectivo (capitalización al cierre en miles de euros).

- Construir una función reciba el fichero de cotizaciones y devuelva un diccionario con los datos del fichero por columnas.

- Construir una función que reciba el diccionario devuelto por la función anterior y cree un fichero en formato csv con el mínimo, el máximo y la media de dada columna.

```python
# Definición del método limpiar() con un argumento
def limpiar(cifra):
    """
    Función que elimina los puntos de separación de miles y cambia las comas de separación de decimales por puntos.
    Parámetros:
        - cifra: Es una cadena con una cifra
    Devuelve:
        Un real con la cifra de la cadena después de eliminar el separador de miles y cambiar el separador de decimales por punto.
    """
    # Se sustituyen los puntos por cadena vacía en el argumento de la función
    cifra = cifra.replace('.', '')
    # Se sustituyen las comas por puntos en el argumento
    cifra = cifra.replace(',','.')
    # Devolvemos el argumento transformado y convertido a número flotante 
    return float(cifra) 

# Definición de la función preprocesado() con un argumento
def preprocesado(ruta):
    """
    Función que preprocesa los datos contenidos en un fichero con formato csv y devuelve 
    un diccionario con los nombres de las columnas como claves y las listas de valores asociados a ellas.
    Parámetros:
        - ruta: Es una cadena con la ruta del fichero.
    Devuelve:
        Un diccionario con pares formados por los nombres de las columnas y las listas de valores en las columnas.
    """
    # Iniciamos bloque try-except
    try:
        # Abrimos el fichero en modo lectura
        with open(ruta, 'r') as f:
            # Leemos el fichero por líneas en una lista
            lineas = f.read().split('\n')
    # Capturamos la excepción Fichero No Encontrado           
    except FileNotFoundError:
        # Informamos por terminal del error
        print('El fichero no existe.')
        # Retorno vacío para terminar el programa
        return
    
    # Sacamos los nombres de las columnas en una lista
    claves = lineas.pop(0).split(";")
    # Creamos el diccionario cotizaciones
    cotizaciones = {}
    # Recorremos la lista de etiquetas
    for i in claves:
        # Creamos una lista vacía por cada clave en el diccionario
        cotizaciones[i] = []
    # Recorremos la lista de líneas
    for linea in lineas:
        # Guardamos en una lista los elementos de cada línea divididos por punto y coma
        campos = linea.split(';')
        # En el diccionario cotizaciones añadimos la primera clave, 
        # y a esa clave le asociamos sólo el primer valor de la lista de campos 
        cotizaciones[claves[0]].append(campos[0])
       # Recorremos con un bucle el rango desde 1 hasta el número de elementos de cada línea
        for i in range(1, len(campos)):
            # Añadimos al diccionario cotizaciones los campos que vamos extrayendo como 
            # listas de datos en el bucle anterior y que transformamos con el 
            # mçetodo limpiar() 
            cotizaciones[claves[i]].append(limpiar(campos[i]))
    # Devolvemos el diccionario 'cotizaciones'        
    return cotizaciones

# Definición de la función resumen_cotizacion() con dos argumentos 
def resumen_cotizacion(cotizaciones, ruta):
    """
    Función que recibe un diccionario con los valores de cotización y crear un fichero 
    con un resumen con el mínimo, el máximo y la media.
    Parámetros:
        - cotizaciones: Es un diccionario con pares cuyas claves son los nombres de la variables medidas y cuyos valores son las listas de valores de cada variable.
        - ruta: Es una cadena con la ruta del fichero.
    """
    # Eliminamos del diccionario pasado por argumento la clave llamada 'Nombre' 
    del(cotizaciones['Nombre'])
    # Creamos una var 'contenido' como cadena vacía
    contenido = ""
    # Añadimos a contenido la cadena 'Nombre'
    contenido += 'Nombre'
    # Recorremos diccionario cotizaciones   
    for i in cotizaciones:
        # Añadimos a la var 'contenido' un punto y coma y un número
        contenido += ";" + i
    # Añadimos a la var 'contenido' un salto de línea seguido de una cadena
    contenido += '\nMínimo'
    # Recorremos los valores del diccionario 'cotizaciones'
    for i in cotizaciones.values():
        # Añadimos el valor mínimo de los valores encontrados
        # a la var 'contenido'
        contenido += ';' + str(min(i))
    # Añadimos a la var 'contenido' un salto de línea seguido de una cadena
    contenido += '\nMáximo'
    # Recorremos los valores del diccionario 'cotizaciones'
    for i in cotizaciones.values():
        # Añadimos el valor máximo de los valores encontrados
        # a la var 'contenido'
        contenido += ';' + str(max(i))
    # Añadimos a la var 'contenido' un salto de línea seguido de una cadena
    contenido += '\nMedia'
    # Recorremos los valores del diccionario 'cotizaciones'
    for i in cotizaciones.values():
        # Añadimos a la var 'contenido' el valor sumatorio de los valores encontrados 
        # divididos por la cantidad de índices
        contenido += ';' + str(sum(i)/len(i))
    # Abrimos fichero pasado como argumento en modo escritura
    with open(ruta, 'w') as f:
        # Escribimos la var 'contenido' en el fichero
        f.write(contenido)
    # Devolvemos en vacío para finalizar el programa
    return


# Guardamos la devoluición al método prepocesado(ruta/al/fichero)
cotizaciones = preprocesado('../data/cotizacion.csv')
# Imprimimos la devolución de prprocesado()
print(cotizaciones)
# Llamamos al método resumen_cotizacion() con el argumento que nos devolvió
# preprocesado()
resumen_cotizacion(cotizaciones, '../data/resumen-cotizacion.csv')
```

```python
def limpiar(cifra):
    return float(cifra.replace('.', '').replace(',', '.'))

def preprocesado(ruta):
    try:
        with open(ruta, 'r') as f:
            lineas = f.read().split('\n')
    except FileNotFoundError:
        print('El fichero no existe.')
        return

    claves = lineas.pop(0).split(";")
    cotizaciones = {i: [] for i in claves}

    for linea in lineas:
        campos = linea.split(';')
        cotizaciones[claves[0]].append(campos[0])
        for i in range(1, len(campos)):
            cotizaciones[claves[i]].append(limpiar(campos[i]))

    return cotizaciones

def resumen_cotizacion(cotizaciones, ruta):
    """
    Crea un fichero con el resumen de cotización (mínimo, máximo y media) de un diccionario de cotizaciones.
    """
    del cotizaciones['Nombre']
    contenido = 'Nombre;' + ';'.join(cotizaciones.keys()) + '\n'
    contenido += 'Mínimo;' + ';'.join(str(min(i)) for i in cotizaciones.values()) + '\n'
    contenido += 'Máximo;' + ';'.join(str(max(i)) for i in cotizaciones.values()) + '\n'
    contenido += 'Media;' + ';'.join(str(sum(i)/len(i)) for i in cotizaciones.values())

    with open(ruta, 'w') as f:
        f.write(contenido)

cotizaciones = preprocesado('../data/cotizacion.csv')
print(cotizaciones)
resumen_cotizacion(cotizaciones, '../data/mi-resumen-cotizacion.csv')
```

7 - El fichero [calificaciones.csv](https://aprendeconalf.es/docencia/python/ejercicios/soluciones/ficheros/calificaciones.csv) contiene las calificaciones de un curso. Durante el curso se realizaron dos exámenes parciales de teoría y un examen de prácticas. Los alumnos que tuvieron menos de 4 en alguno de estos exámenes pudieron repetirlo en la al final del curso (convocatoria ordinaria). Escribir un programa que contenga las siguientes funciones:

Una función que reciba el fichero de calificaciones y devuelva una lista de diccionarios, donde cada diccionario contiene la información de los exámenes y la asistencia de un alumno. La lista tiene que estar ordenada por apellidos.

Una función que reciba una lista de diccionarios como la que devuelve la función anterior y añada a cada diccionario un nuevo par con la nota final del curso. El peso de cada parcial de teoría en la nota final es de un 30% mientras que el peso del examen de prácticas es de un 40%.

Una función que reciba una lista de diccionarios como la que devuelve la función anterior y devuelva dos listas, una con los alumnos aprobados y otra con los alumnos suspensos. Para aprobar el curso, la asistencia tiene que ser mayor o igual que el 75%, la nota de los exámenes parciales y de prácticas mayor o igual que 4 y la nota final mayor o igual que 5.

```python
# Definición de la función 'nota()' con un argumento
def nota(cifra):
    '''Función que elimina cambia las comas de separación de decimales por puntos de una cifra y la convierte en un real.
    Parámetros:
        - cifra: Es una cadena con una cifra.
    Devuelve:
        Un real con la cifra de la cadena después de cambiar el separador de decimales por punto.
    '''
    # Se sustituyen las comas por puntos en el argumento    
    cifra = cifra.replace(',','.')
    # Devolvemos el argumento transformado y convertido a número flotante 
    return float(cifra) 

def calificaciones(ruta):
    '''Función que preprocesa los datos contenidos en un fichero con formato csv y devuelve una lista de diccionarios donde las claves de los diccionarios son los datos de la primera fila y los valores los datos de cada línea del fichero.
    
    Parámetros:
        - ruta: Es una cadena con la ruta del fichero.
    Devuelve:
        Una lista de diccionarios donde cada diccionario contiene los datos de una linea del fichero (a excepción de la primera línea), usando como claves los datos de la primera línea.
    '''
    try:
        # Abrimos el fichero en modo lectura
        f = open(ruta, 'r')
    # Capturamos la excepción 'Fichero No Encontrado'
    except FileNotFoundError:
        # Imprimimos el error
        print('El fichero no existe.')
        # Devolución vacía para terminar el programa
        return
    # En una lista guardamos las líneas del fichero
    lineas = f.readlines()
    # Cerramos el fichero
    f.close()
    # Guardamos los elementos de la primera línea separados por punto y coma en una lista 
    # menos el último '[:-1]' porque es un salto de línea
    claves = lineas[0][:-1].split(";")
    # Creamos la lista vacía 'calificaciones'
    calificaciones = []
    # Recorremos mediante bucle la lista de líneas 
    for i in lineas[1:]:
        # Creamos una lista con los elementos de cada línea, separando por los puntos y coma
        valores = i[:-1].split(";")
        # Creamos el diccionario vacío 'alumno'
        alumno = {}
        # Recorremos mediante bucle el rango 0 hasta el número de elementos de la lista
        # recíen creada con los elementos de la línea 
        for j in range(len(valores)):
            # Añadimos al diccionario 'alumno' el índice 'j' como clave
            # y la lista de valores recorrida
            alumno[claves[j]] = valores[j]
        # Añadimos a la lista 'calificaciones' el diccionario 'alumno' recién creado
        calificaciones.append(alumno)
    # Devolvemos la lista 'calificaciones' con todos los diccionarios 'alumno' 
    return calificaciones

# Definición de la función 'añadir_nota_final(calificaciones)' con un argumento
def añadir_nota_final(calificaciones):
    '''Función que recibe una lista de diccionarios con las calificaciones de cada alumno en un curso, calcula la nota final del curso de cada alumno y la añade al diccionario del alumno.

    Parámetros:
        - calificaciones: Es una lista de diccionarios donde cada diccionario contiene los datos de un alumno (nombre, asistencia y notas de exámenes del curso).
    Devuelve:
        La lista de las calificaciones de los alumnos tras añadir a cada alumno de la lista su nota final del curso.
    '''
    # Definición de función privada 'nota_final(alumno)' con un argumento
    def nota_final(alumno):
        '''Función que calcula la nota final del curso de un alumno.

        Parámetros:
            - alumno: Es un diccionario con las notas del alumno.
        Devuelve:
            El diccionario con los datos del alumno tras añadirle un nuevo par con la nota final del curso.
        '''
        
        if alumno['Ordinario1']: #Si el alumno se ha presentado al examen de repesca del primer parcial tomamos esa nota como la nota del primer parcial
            # Guardamos la nota en una variable después de transformarla 
            parcial1 = nota(alumno['Ordinario1'])
        # #Si el alumno se ha presentado al examen del primer parcial
        elif alumno['Parcial1']:
            # Guardamos la nota en una variable después de transformarla 
            parcial1 = nota(alumno['Parcial1'])
        else: # No se ha presentado al primer parcial ni a la repesca en el ordinario
            # Asignamos valor '0' al primer parcial
            parcial1 = 0
        if alumno['Ordinario2']: #Si el alumno se ha presentado al examen de repesca del segundo parcial tomamos esa nota como la nota del segundo parcial
            parcial2 = nota(alumno['Ordinario2'])
        # Si el alumno se ha presentado al examen del segundo parcial
        elif alumno['Parcial2']:
            # Guardamos la nota en una variable después de transformarla 
            parcial2 = nota(alumno['Parcial2'])
        else: # No se ha presentado al segundo parcial ni a la repesca en el ordinario
            # Asignamos valor '0' al segundo parcial 
            parcial2 = 0 
        if alumno['OrdinarioPracticas']: #Si el alumno se ha presentado al examen de repesca de prácticas tomamos esa nota como la nota de prácticas
            # Guardamos la nota en una variable después de transformarla 
            practicas = nota(alumno['OrdinarioPracticas'])
        # Si el alumno se ha presentado al examen de prácticas
        elif alumno['Practicas']:
            # Guardamos la nota en una variable después de transformarla 
            practicas = nota(alumno['Practicas'])
        # Si no se cumplen las dos condiciones
        else:
            # Asignamos valor '0' a las prácticas
            practicas = 0
        # Asignamos el valor del primer parcial al diccionario 'alumno' en la clave del primer final
        alumno['Final1'] = parcial1
        # Asignamos el valor del segundo parcial al diccionario 'alumno' en la clave del segundo final
        alumno['Final2'] = parcial2
        # Asignamos el valor de las prácticas al diccionario 'alumno' en la clave del final de prácticas
        alumno['FinalPracticas'] = practicas
        # Asignamos el valor de la nota final como una suma ponderada de los parciales y las prácticas
        alumno['NotaFinal'] = parcial1 * 0.3 + parcial2 * 0.3 + practicas * 0.4
        # Devolvemos el diccionario 'alumno'
        return alumno
    # Devolvemos una lista en la que se aplica la función 'nota_final()' a cada elemento de la lista calificaciones
    return list(map(nota_final, calificaciones))

# Definición de la función 'aprobados_suspensos(calificaciones)' con un argumento
def aprobados_suspensos(calificaciones):
    '''Función que recibe una lista de diccionarios con las calificaciones de cada alumno en un curso, y devuelve la lista de aprobados y suspensos en el curso.

    Parámetros:
        - calificaciones: Es una lista de diccionarios donde cada diccionario contiene los datos de un alumno (nombre, asistencia y notas de exámenes del curso).
    Devuelve:
        - aprobados: Es una lista con los nombres de los alumnos aprobados.
        - suspensos: Es una lista con los nombres de los alumnos suspensos.
    '''
    # Creamos dos listas vacías
    aprobados = []
    suspensos = []
    # Recorremos lista calificaciones
    for alumno in calificaciones:
        # Si se cumplen todas las condiciones
        if all([int(alumno['Asistencia'][:-1]) >= 75, alumno['Final1'] >= 4, alumno['Final2'] >=4, alumno['FinalPracticas'] >=4, alumno['NotaFinal'] >= 5]):
            # Añadimos los apellidos y el nombre del alumno a la lista aprobados
            aprobados.append(alumno['Apellidos'] + ', ' + alumno['Nombre'])
        # Si no se cumplen todas
        else:
            # Añadimos los apellidos y el nombre del alumno a la lista suspensos
            suspensos.append(alumno['Apellidos'] + ', ' + alumno['Nombre'])
    # Devolvemos las listas aprobados y suspensos
    return aprobados, suspensos

# Imprimimos la devolución de la función 'añadir_nota_final()' que recibe como argumento
#  lo que devuelve la función 'calificaciones()'
print(añadir_nota_final(calificaciones('../data/calificaciones.csv')))
# Recogemos en dos listas los aprobados y los suspensos
# llamando a la función 'aprobados_suspensos()' 
aprobados, suspensos = aprobados_suspensos(añadir_nota_final(calificaciones('../data/calificaciones.csv')))
# Imprimimos la lista de aprobados y la de suspensos
print('Lista de aprobados:', aprobados)
print('Lista de suspensos:', suspensos)
```
